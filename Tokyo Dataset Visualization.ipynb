```python
    
    import pandas as pd #For data manipulation and analysis
    import matplotlib.pyplot as plt #For visualizing the data
    import numpy as np #For performing mathematical operations
    import os #For managing the directory/folder

    #Read the Excel file from the system
    for dirname, _, filenames in os.walk('C:\\Users\\HP\\OneDrive\\Desktop\\Olympics in Tokyo Dataset'):
        for filename in filenames:
            gender = pd.read_excel(r'C:\\Users\\HP\\OneDrive\\Desktop\\Olympics in Tokyo Dataset\\Gender.xlsx')

    #Calculate the gender counts
    m,f = 0,0
    for x in range(len(gender['Discipline'])):
        f+=gender['Female'][x]
        m+=gender['Male'][x]

    #Plot the Excel data
    plt.title('FEMALE AND MALE ATHLETE DISTRIBUTION', fontsize=20)
    flbl = 'Female Athletes '+ str(f*100/(f+m))+'%'
    mlbl = 'Male Athletes '+str(m*100/(f+m))+'%'
    plt.pie([f,m],labels=[flbl, mlbl], radius=1)
    plt.show()
    
```

```python
    
    import pandas as pd #For data manipulation and analysis
    import matplotlib.pyplot as plt #For visualizing the data
    import numpy as np #For performing mathematical operations
    import os #For managing the directory/folder

    #Read the Excel file from the system
    for dirname, _, filenames in os.walk('C:\\Users\\HP\\OneDrive\\Desktop\\Olympics in Tokyo Dataset'):
        for filename in filenames:
            athlete = pd.read_excel(r'C:\\Users\\HP\\OneDrive\\Desktop\\Olympics in Tokyo Dataset\\Athletes.xlsx')

    #Calculate the events
    l=[]
    for x in list(set(athlete['Discipline'])):
        l.append(list(athlete['Discipline']).count(x))

    #Plot the Excel data
    plt.rcParams["figure.figsize"] = (14, 5) #Runtime Configuration (rc) contain default styles for plot element
    plt.bar(list(set(athlete['Discipline'])),l)
    plt.title('EVENT WISE DISTRIBUTION OF PLAYERS', fontsize=20)
    plt.xlabel('Events')
    plt.ylabel('Number of athletes')
    plt.xticks(rotation=90) #Get and set the locations and labels of x-axis
    plt.show()
    
```
